```python
import requests
from bs4 import BeautifulSoup
import ollama
from tqdm import tqdm
import time
```

```python
def get_webpage_text(url):
    print(f"üåê Fetching URL: {url}")
    response = requests.get(url)
    if response.status_code != 200:
        raise Exception(f"‚ùå Failed to load page: {response.status_code}")
    soup = BeautifulSoup(response.text, 'html.parser')
    paragraphs = soup.find_all('p')
    text = '\n'.join([p.get_text() for p in paragraphs])
    print(f"üßæ Extracted {len(paragraphs)} paragraphs.")
    return text
```

```python


def summarize_text(text):
    prompt = f"Please summarize the following text:\n\n{text}\n\nSummary:"
    print("‚úÖ Starting chat with Ollama...")

    try:
        print("üì§ Sending request to model 'qwen3'...")
        for _ in tqdm(range(5), desc="‚åõ Waiting for model response"):
            time.sleep(0.5)  # Fake progress while waiting

        response = ollama.chat(
            model='qwen3',
            messages=[{
                'role': 'user',
                'content': prompt
            }]
        )

        print("üì• Received response from model.")

        # Show raw response
        print("üßæ Raw response:")
        print(response)

        if 'message' in response and 'content' in response['message']:
            content = response['message']['content']
            print("\nüìù Full message content:\n")
            print(content)

            print("\nüî† Word-by-word output:\n")
            for word in content.split():
                print(word)

            return content
        else:
            print("‚ö†Ô∏è Unexpected response structure.")
            return None

    except Exception as e:
        print("‚ùå An error occurred:")
        print(e)
        return None

# Example usage
url = "https://en.wikipedia.org/wiki/Natural_language_processing"
webpage_text = get_webpage_text(url)
summary = summarize_text(webpage_text[:1000])  # limit for performance
```
